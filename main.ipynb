Import Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing Datasets

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

Training the Random Forest Regression model on the whole datasets

```python
from sklearn.ensemble import RandomForestRegressor

rfr_regressor = RandomForestRegressor(n_estimators=10, random_state=0)
rfr_regressor.fit(X, y)
```

Single Value Predict

```python
y_predict = rfr_regressor.predict([[6.5]])
y_predict
```

Multiple Value Predict

```python
y_predict = rfr_regressor.predict(X)
y_predict
```

Visualizing the Random Forest Regression

```python
plt.scatter(X, y, color='red')
plt.plot(X, rfr_regressor.predict(X), color='blue')
plt.title('Salary Expectation')
plt.xlabel('Employee Level')
plt.ylabel('Salary')
plt.show()
```

Higher Resolution of Random Forest Regression

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((-1, 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, rfr_regressor.predict(X_grid), color='blue')
plt.title('Salary Expectation')
plt.xlabel('Employee Level')
plt.ylabel('Salary')
plt.show()
```

```python
rfr_regressor.predict(X_grid)
```

Prediction and Evaluate

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Metrics Calculation
mse = mean_squared_error(y, y_predict)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_predict)
r2 = r2_score(y, y_predict)

# Print Results
print(f'MSE: {mse:.4f}')
print(f'RMSE: {rmse:.4f}')
print(f'MAE: {mae:.4f}')
print(f'RÂ² Score: {r2:.4f}')
```
